```python
class Counter:
    def __init__(self,limit,initial=0,min_digits = 1):
        self.limit = limit
        self.min_digits = min_digits
        
        if initial >= 0 and initial <= self.limit - 1:
            self.initial = initial
        else:
            print('Error')
            self.initial = self.limit - 1
            
    def get_value(self):
        return self.initial
    
    def __str__(self):
        length = len(str(self.initial))
        return "0" * (self.min_digits - length) + str(self.initial)
    
    def tick(self):
        if self.initial == 0:
            self.initial = self.limit - 1
            return True
        else:
            self.initial -= 1
            return False
```

```python
counterA = Counter(60,5,4)
counterA.get_value()
print(counterA)
print()
counterA.tick()
print(counterA)
counterA.tick()
```

```python
class Timer:
    def __init__(self,hours,minutes,seconds):
        self.hours = Counter(24,hours,2)
        self.minutes = Counter(60,minutes,2)
        self.seconds = Counter(60,seconds,2)
        
    def __str__(self):
        return self.hours.__str__() + ":" + self.minutes.__str__() + ":" + self.seconds.__str__()
    
    def tick(self):
        if self.seconds.tick():
            if self.minutes.tick():
                self.hours.tick()
                
    def is_zero(self):
        return str(self.hours) == "00" and str(self.minutes) == "00" and str(self.seconds) == "00"
```

```python
timerA = Timer(1,30,0)
while True:
    print(timerA)
    timerA.tick()
    if timerA.is_zero():
        print(timerA)
        break
```
